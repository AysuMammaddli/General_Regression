```python
def target_corr(data, target, threshold=0.6):
    corr_matrix=data.corr(method='spearman', numeric_only=True)
    target_correlation=corr_matrix[target].abs()>threshold
    var_exp_target=target_correlation[target_correlation].index.tolist()
    var_exp_target.remove(target)

    result=pd.DataFrame({
    'variable':var_exp_target,
    'correlation':corr_matrix.loc[var_exp_target, target]
  })
    result.reset_index(drop=True, inplace=True)
    return result
```

```python
target=target_corr(data_lr, 'Life expectancy ')
target
```

```python
def intercorr(data, target, threshold=0.8):
    corr_matrix=data.drop(columns=[target]).corr(method='spearman', numeric_only=True)
    high_corr_var=(corr_matrix.abs()>threshold) & (corr_matrix!=1)
    independent_var=np.where(high_corr_var)
    result=pd.DataFrame({
        'variable1':corr_matrix.index[independent_var[0]],
        'variable2':corr_matrix.columns[independent_var[1]]
    })

    return result
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
def vif_result(data, variables):
    vif_data=data[variables]
    vif_values=[variance_inflation_factor(vif_data.values, i) for i in range(vif_data.shape[1])]
    vif=pd.DataFrame({'VIF':vif_values,
                    'Feature':variables})
    return vif
```

```python
from sklearn.linear_model import LinearRegression
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from catboost import CatBoostRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
```

```python
models=[]

LR_def=LinearRegression()
RF_def=RandomForestRegressor()
XGB_def=XGBRegressor()
LGBM_def=LGBMRegressor()
CB_def=CatBoostRegressor()
CB_cat_def=CatBoostRegressor(cat_features=['Status'])
SVR_def=SVR()

models.extend([
    ('LinearRegression_default', LR_def),
              ('RandomForest_default', RF_def),
              ('XGBoost_default', XGB_def),
              ('LightGBM_default', LGBM_def),
              ('CatBoost_default', CB_def),
              ('CatBoost_categoric_default', CB_cat_def),
              ('SVR_default', SVR_def)
])
```

```python
from sklearn import metrics
def train_and_evaluate(model_name, model, x_train, y_train, x_test, y_test):
    y_train=y_train.fillna(value=y_train.mean())
    y_test=y_test.fillna(value=y_test.mean())
    model.fit(x_train, y_train)
    y_pred=model.predict(x_test)
    r2_test=metrics.r2_score(y_test, y_pred)

    print(f'Model performance is {model_name}')
    print(f'r2 is {r2_test}')
    return  r2_test
```

```python
r2_df=pd.DataFrame(columns=['model', 'r2_score'])


for model_name, model in models:
    if model_name=='LinearRegression_default':
        r2=train_and_evaluate(model_name, model, x_train_lr, y_train_lr, x_test_lr, y_test_lr)
    elif model_name=='CatBoost_categoric_default':
        r2=train_and_evaluate(model_name, model, x_train_cat, y_train_cat, x_test_cat, y_test_cat)
    elif model_name in ['RandomForest_default', 'SVR_default']:
        x_train_filled=x_train_dummy.copy()
        x_test_filled=x_test_dummy.copy()
        for i in x_train_filled.columns:
            if x_train_filled[i].dtypes==object:
                x_train_filled[i]=x_train_filled[i].fillna(value=x_train_filled[i].mode()[0])
                x_test_filled[i]=x_test_filled[i].fillna(value=x_test_filled[i].mode()[0])
            else:
                x_train_filled[i]=x_train_filled[i].fillna(value=x_train_filled[i].mean())
                x_test_filled[i]=x_test_filled[i].fillna(value=x_test_filled[i].mean())
        r2=train_and_evaluate(model_name, model, x_train_filled, y_train_dummy, x_test_filled, y_test_dummy)
    else:
        r2=train_and_evaluate(model_name, model, x_train_dummy, y_train_dummy, x_test_dummy, y_test_dummy)
    if r2 is not None:
        r2_df=pd.concat([r2_df, pd.DataFrame({'model':[model_name], 'r2_score':[r2]})], ignore_index=True)

r2_df_sorted=r2_df.sort_values(by='r2_score', ascending=False)
r2_df_sorted
```
